# Chess Tutor Web App – Tutorial

## Overview

This project is an interactive chess web application designed to help players improve their understanding of the game through real-time feedback. Unlike traditional chess engines that only provide optimal moves, this system combines engine analysis with natural language explanations to guide users toward better decision-making.

The platform integrates:
- A playable chess board with real-time validation
- Engine-backed analysis (Stockfish + Maia)
- A tutoring system that explains moves and suggests improvements
- Configurable bots and game settings

This tutorial walks through the user interface and explains how to effectively use each component.

## Main Board, Clocks, and FEN

The central component of the UI is the chess board. This is where all interaction happens.

![Board](assets/board.png)

### Board Interaction
- Click and drag pieces to make moves
- Only legal moves are allowed
- The board updates immediately after each move

### Clocks
- Each player has a clock displayed near the board
- Time decreases automatically during play
- Clock settings are configurable (see New Game panel)

### FEN Display
- The current board position is represented as a FEN string
- Useful for:
  - Debugging
  - Reproducing positions
  - External engine analysis

The board, clocks, and FEN are tightly synchronized and represent the current game state at all times.

## Left Panel Controls (Engine, Bot, Flip Board)

The left-side panel provides controls that modify how the system behaves.

![Left Panel](assets/left_panel.png)

### Engine Toggle
- Enables or disables engine evaluation
- When enabled:
  - The system computes best moves
  - Tutor feedback becomes more informed
- When disabled:
  - No engine analysis is performed

### Bot Toggle
- Enables playing against a bot
- When active:
  - The system automatically responds with moves
  - Bot strength depends on configured ELO

### Flip Board
- Reverses board orientation
- Useful for:
  - Playing as Black
  - Analyzing positions from the opponent’s perspective

### Interaction Between Controls
- Engine + Bot ON → full experience (play + feedback)
- Engine ON, Bot OFF → analysis mode
- Engine OFF, Bot ON → bot plays without deep evaluation (limited usefulness)

These controls dynamically affect how the backend processes moves and generates responses.

## New Game Panel

The New Game panel allows you to configure and start a new session.

![New Game](assets/new_game.png)

### Configurable Fields

#### Clock Settings
- Set time controls for each player
- Typically includes:
  - Initial time (e.g., 5 minutes)
  - Increment (optional)

#### ELO Setting
- Controls bot strength
- Lower ELO → weaker, more human-like play
- Higher ELO → stronger, more engine-like play

### New Game Button
- Initializes a fresh game with selected settings
- Resets:
  - Board state
  - Clocks
  - Move history
  - Tutor context

This panel is the primary way to customize gameplay conditions.

## Tutor Panel

The tutor panel provides real-time feedback and guidance.

![Tutor](assets/tutor.png)

### Features

- **Move Feedback**
  - Evaluates your last move
  - Labels it (e.g., good move, mistake, blunder)

- **Suggestions**
  - Recommends better alternatives
  - Often includes reasoning

- **Positional Insights**
  - Highlights concepts such as:
    - King safety
    - Piece activity
    - Tactical threats

- **Encouragement**
  - Provides human-like coaching feedback

### How It Works

The tutor system:
1. Receives engine evaluations
2. Compares your move to optimal lines
3. Generates natural language explanations

This is the core feature that transforms the application from a chess engine into a learning tool.

## Conclusion

This application combines gameplay, analysis, and instruction into a single interface. By integrating engine evaluations with human-readable feedback, it provides a more effective way to learn chess compared to traditional tools.

Key takeaways:
- Use the board to play and explore positions
- Adjust settings in the New Game panel to match your skill level
- Enable the engine and tutor for maximum learning value
- Experiment with different configurations to understand how gameplay changes

The system is designed to be both interactive and instructional, making it suitable for players looking to actively improve.
