# Chess Tutor Web App – Tutorial

## Overview

This project is an interactive chess web application designed to help players improve their understanding of the game through real-time feedback. Unlike traditional chess engines that only provide optimal moves, this system combines engine analysis with natural language explanations to guide users toward better decision-making.

The platform integrates:
- A playable chess board with real-time move validation
- Engine-backed analysis (Stockfish + Maia) with an evaluation bar and best-move arrow
- A tutoring system that explains every player move and supports follow-up questions
- Novice Mode for beginner-friendly explanations
- Configurable bots and game settings that persist across browser refreshes

This tutorial walks through the user interface and explains how to effectively use each component.


## Main Board, Clocks, and FEN

The central component of the UI is the chess board. This is where all interaction happens.

![Board](assets/board.png)

### Board Interaction
- Click and drag pieces to make moves
- Only legal moves are allowed
- The board updates immediately after each move

### Clocks
- Each player has a clock displayed near the board
- Time decreases automatically during play
- Clock settings are configurable (see New Game panel)

### FEN Display
- The current board position is represented as a FEN string
- Useful for:
  - Debugging
  - Reproducing positions
  - External engine analysis

The board, clocks, and FEN are tightly synchronized and represent the current game state at all times.

## Left Panel Controls

The left-side panel provides controls that modify how the system behaves during play.

### Engine Toggle
- Enables or disables Stockfish evaluation
- When enabled:
  - The evaluation bar appears to the left of the board, showing who is winning
  - A best-move arrow is drawn on the board
  - The Engine panel (right side) shows depth, eval score, and the principal variation
- When disabled: eval bar and arrow are hidden

### Bot Toggle
- Enables playing against the AI opponent
- When active, the system automatically responds with moves after yours
- Bot strength is set by the ELO field in the New Game panel
- The bot uses a blend of Maia (human-like play) and Stockfish, producing realistic mistakes

### Tutor Toggle
- Enables move-by-move coaching feedback after each of your moves
- When on, the Tutor panel shows a plain-prose explanation: why your move was good or bad, what the engine preferred, and key tactical details
- The On/Off badge next to the Tutor heading is also clickable as a shortcut
- A follow-up question field appears after each explanation so you can ask for clarification

### Flip Board
- Mirrors the board orientation without resetting the game
- Useful for reviewing positions from the opponent's perspective

### Toggle Persistence
All toggle states (Engine, Bot, Tutor, Novice Mode) and the ELO value are automatically saved to browser localStorage. If you refresh the page, every toggle is restored to its previous state — you do not need to re-enable anything manually.


## New Game Panel

The New Game panel allows you to configure and start a new session.

### Configurable Fields

#### Clock Time
- Sets the time control for both players
- Options: No Clock, 1 min, 3 min, 5 min, 10 min
- When No Clock is selected, time does not count down

#### Color
- Choose to play as White, Black, or Random
- Selecting Black automatically flips the board so Black is on the bottom

#### ELO
- Sets the bot's target strength (0–3000)
- Lower ELO → weaker, more human-like play with more mistakes
- Higher ELO → stronger, fewer errors

| Level | ELO Range |
|---|---|
| Beginner | 100 – 800 |
| Amateur | 800 – 1400 |
| Intermediate | 1400 – 1800 |
| Advanced | 1800 – 2200 |
| Expert / Master | 2200+ |
| Grandmaster | 2500 |
| Super GM | 2700 |
| Engine-level | 3000+ |

### New Game Button
- Starts a fresh game with the selected settings
- Resets: board state, clocks, move history, tutor conversation context
- If the Bot and/or Engine toggles are on, they restart automatically with the new settings


## Tutor Panel

The Tutor panel (right side of the screen) provides real-time coaching feedback after each of your moves.

### Enabling the Tutor
Click the **Tutor** toggle in the left panel, or click the **On/Off badge** next to the Tutor heading in the right panel. Both controls toggle the same setting.

### What the Tutor Explains
After each player move, the tutor:
1. Runs Stockfish at depth 12 on the position before and after the move
2. Computes the centipawn loss (how much the move cost relative to the best alternative)
3. Classifies the move:
   - **Fine** — ≤20 cp lost
   - **Inaccuracy** — 21–50 cp lost
   - **Mistake** — 51–100 cp lost
   - **Blunder** — >100 cp lost
4. Computes tactical facts from python-chess: what was captured, whether the destination was under attack, which pieces are left undefended, and whether check was given
5. Passes all verified data to an LLM, which produces a plain-prose explanation

The tutor only runs on **your moves**, not the bot's moves.

### Follow-Up Questions
After the tutor displays an explanation, a text field appears beneath it. You can type any question, such as:
- "What do you mean by that?"
- "What if I had moved my knight to f6 instead?"
- "Can you explain that more simply?"

Submit with Enter or the Ask button. The follow-up has access to all the same position data as the original explanation. You can ask multiple follow-ups in sequence. The context resets when you make your next move.

### Novice Mode
The **Novice Mode** chip next to the Tutor heading instructs the tutor to use beginner-friendly language and define any chess terms it uses. Enable it if you are unfamiliar with terms like "en prise," "fork," or "principal variation."

### Tips for Getting the Most Out of the Tutor
- Read every explanation, even for moves classified as fine
- Use follow-up questions to ask about anything unclear
- Enable Novice Mode if terminology is confusing
- Pay extra attention to blunders and mistakes
- Use the Engine toggle alongside the Tutor to see the eval bar shift as you make decisions
